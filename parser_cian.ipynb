```python
#connecting the library for parsing
#and for create http request
from bs4 import BeautifulSoup
import requests
import string
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#link parsing
def parse_links(url):
  
  request = requests.get(url)
  bs = BeautifulSoup(request.text, 'html.parser')

  #finding the block
  all_links = bs.find_all('a', class_ = '_93444fe79c--link--eoxce')

  #creating an array of links
  link_href = []
  for link in all_links:
    link_href.append(link['href'])
  return link_href
```

```python
def price(url):
  url = "https://www.cian.ru/kupit-kvartiru-moskva-moskovskiy/"

  request = requests.get(url)

  bs = BeautifulSoup(request.text, "html.parser")

  #finding the block
  all_links = bs.find_all("p", class_ = "_93444fe79c--color_gray60_100--mYFjS _93444fe79c--lineHeight_20px--fX7_V _93444fe79c--fontWeight_normal--JEG_c _93444fe79c--fontSize_14px--reQMB _93444fe79c--display_block--KYb25 _93444fe79c--text--e4SBY _93444fe79c--text_letterSpacing__normal--tfToq")
  return all_links
```

```python
#Region: 
#Moscow oblast: '-moskva-moskovskiy/';
```

```python
#Type of real estate: 
#buy an apartment: 'kupit-kvartiru'
```

```python
#we enter the region here for reference
region = input()
```

```python
#we enter the type of real estate here for reference
type = input()
```

```python
#конструируем ссылку
url = 'https://www.cian.ru/' + type + region
```

```python
cian_links = parse_links(url)
```

```python
data = price(url)

x = 28
data_price = []
for i in range(x):
    str_data = str(data[i])
    data_p = str_data[-16:-9]
    data_pr = data_p[0:3]+data_p[4:]
    int_data = int(data_pr)
    data_price.append(int_data)

df = pd.Series(data_price)
```

```python
plt.hist(df)
plt.xlabel('price per M^2')
plt.ylabel('frequency')
plt.title('price frequency')
plt.show()
```

```python
plt.boxplot(df)
plt.title('price per M^2')
plt.show()
```

```python
sns.displot(data=df, kde=True)
plt.title("the probability of the price")
plt.show()
```

```python
print("Descriptive statistics")
print("Mean: ", np.mean(df))
print("Median: ", np.median(df))
print("Variance: ", np.var(df))
print("Standard deviation: ", np.std(df))
```
